```python
import os
import glob
import random
import shutil

# ============================================================
# CHEMIN DATASET
# ============================================================
DATASET_DIR = "/content/drive/MyDrive/dislocation_2026/dataset_classification"

# ============================================================
# RATIOS
# ============================================================
TRAIN_RATIO = 0.7
VAL_RATIO   = 0.15
TEST_RATIO  = 0.15

# ============================================================
# CLASSES
# ============================================================
CLASSES = ["blanc", "noir", "mixte"]

# ============================================================
# OUTPUT
# ============================================================
OUTPUT_SPLIT = os.path.join(DATASET_DIR, "split")

# ============================================================
# SUPPORTED EXTENSIONS
# ============================================================
EXTENSIONS = ["*.png", "*.jpg", "*.jpeg", "*.tif", "*.tiff"]

# ============================================================
# SUPPRIMER L'ANCIEN SPLIT S'IL EXISTE (optionnel)
# ============================================================
if os.path.exists(OUTPUT_SPLIT):
    print(f"🗑️ Suppression de l'ancien dossier split...")
    shutil.rmtree(OUTPUT_SPLIT)

# ============================================================
# CREATION DOSSIERS
# ============================================================
for split in ["train", "val", "test"]:
    for c in CLASSES:
        os.makedirs(os.path.join(OUTPUT_SPLIT, split, c), exist_ok=True)

# ============================================================
# LOAD IMAGES (multi-format)
# ============================================================
def load_images(folder):
    files = []
    for ext in EXTENSIONS:
        files.extend(glob.glob(os.path.join(folder, ext)))
    return files

# ============================================================
# SPLIT FUNCTION
# ============================================================
def split_class(class_name):

    folder = os.path.join(DATASET_DIR, class_name)
    images = load_images(folder)

    if len(images) == 0:
        print(f"\n⚠️ Classe {class_name} vide ou introuvable")
        return

    random.shuffle(images)

    n_total = len(images)
    n_train = int(n_total * TRAIN_RATIO)
    n_val   = int(n_total * VAL_RATIO)

    train_files = images[:n_train]
    val_files   = images[n_train:n_train + n_val]
    test_files  = images[n_train + n_val:]

    print(f"\n📂 Classe {class_name}")
    print(f"   total : {n_total}")
    print(f"   train : {len(train_files)}")
    print(f"   val   : {len(val_files)}")
    print(f"   test  : {len(test_files)}")

    # ========================================================
    # COPY
    # ========================================================
    def copy_files(files, split):
        for f in files:
            name = os.path.basename(f)
            dst = os.path.join(OUTPUT_SPLIT, split, class_name, name)
            shutil.copy2(f, dst)

    copy_files(train_files, "train")
    copy_files(val_files, "val")
    copy_files(test_files, "test")

# ============================================================
# MAIN
# ============================================================
print("🚀 SPLIT DATASET START")
print(f"📂 Dataset : {DATASET_DIR}")
print(f"📁 Output : {OUTPUT_SPLIT}")
print("="*50)

for c in CLASSES:
    split_class(c)

print("\n" + "="*50)
print("✅ SPLIT TERMINÉ")
print(f"📁 Résultat : {OUTPUT_SPLIT}")

# ============================================================
# VÉRIFICATION FINALE
# ============================================================
print("\n📊 RÉSUMÉ FINAL :")
print("-"*50)
for split in ["train", "val", "test"]:
    print(f"\n{split.upper()} :")
    for c in CLASSES:
        split_dir = os.path.join(OUTPUT_SPLIT, split, c)
        if os.path.exists(split_dir):
            n_images = len(os.listdir(split_dir))
            print(f"   {c}: {n_images} images")
```
